# Question 1: What is hypothesis testing in statistics?

--> Hypothesis testing is a formal statistical procedure that uses sample data to evaluate a claim or assumption about a population. It is a structured process that helps researchers determine whether observed results are likely to have occurred by chance or if they represent a genuine effect.

# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

--> Null Hypothesis: A statement of no effect, no difference, or no relationship between variables. It is the default assumption, assumed to be true until evidence suggests otherwise (Ex: the average weight of men is not different from the average weight of women).

--> Alternative Hypothesis: A statement that contradicts the null hypothesis, proposing that an effect, difference, or relationship does exist. This is usually the claim the researcher is trying to support with evidence (Ex: the average weight of men is greater than the average weight of women). 

# Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

--> The significance level is the probability of rejecting a null hypothesis that is actually true. It's a measure of how strong the sample evidence must be to conclude that the results are not due to random chance.

-->The significance level acts as the benchmark for statistical significance. The outcome of a hypothesis test is determined by comparing the calculated p-value to the chosen significance level (alpha).

If p-value < alpha : The result is considered statistically significant. You reject the null hypothesis.

If p-value > alpha : The result is not statistically significant. You fail to reject the null hypothesis. 

# Question 4: What are Type I and Type II errors? Give examples of each.

--> A Type I error is a false positive, where a true null hypothesis is incorrectly rejected (Ex: a covid positive person is wrongly told he isn't positive). 

--> A Type II error is a false negative, where a false null hypothesis is incorrectly accepted (Ex: a covid negative person is wrongly told he is positive). 
  

# Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

--> Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between the means of two samples. It is typically used when the sample size is large (generally n > 30) and the population standard deviation is known. The Z-test is based on the standard normal distribution (Z-distribution).

The Z-test compares the means of two populations with a large sample size (typically ≥ 30) and known population standard deviation. It assesses whether the difference between the means is statistically significant.

--> T-test is a statistical test used to determine whether there is a significant difference between the means of two groups.

It is particularly useful when the sample size is small (typically n < 30) and the population standard deviation is unknown. The T-test relies on the t-distribution, which is similar to the normal distribution but has heavier tails.

# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

>sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


mu = 50.0 
sigma = 1.5 
alpha = 0.05 

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_statistic = (sample_mean - mu) / (sigma / np.sqrt(sample_size))

z_critical = norm.ppf(1 - alpha/2)

p_value = 2 * norm.sf(abs(z_statistic))


print(f"Sample Mean: {sample_mean}")
print(f"Sample Size: {sample_size}")
print(f"Z-Statistic: {z_statistic}")
print(f"Critical Z-Value: {z_critical}")
print(f"P-Value: {p_value}")

if abs(z_statistic) > z_critical:
    print("Result: Reject the Null Hypothesis (H0)")
else:
    print("Result: Fail to reject the Null Hypothesis (H0)")

if p_value < alpha:
    print("Conclusion: The sample mean is statistically different from the hypothesized population mean (p < 0.05).")
else:
    print("Conclusion: The sample mean is not statistically different from the hypothesized population mean (p >= 0.05).")

```

# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 50   
sigma = 10 
n = 100  

data = np.random.normal(loc=mu, scale=sigma, size=n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

z_critical = norm.ppf(0.975)
margin_of_error = z_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean}")
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
print(f"True Population Mean (for reference): {mu}")

plt.hist(data, bins=15, density=True)

plt.title(f'Normal Distribution Simulation and 95% Confidence Interval (N={n})')
plt.xlabel('Value')
plt.ylabel('Density')

```

# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_z_scores(data):

    mean_val = np.mean(data)
    std_val = np.std(data)

    z_scores = (data - mean_val) / std_val
    return z_scores

np.random.seed(42)
sample_data = np.random.normal(loc=70, scale=10, size=200)

standardized_data = calculate_z_scores(sample_data)

print(f"Original Mean: {np.mean(sample_data)}")
print(f"Original Std Dev: {np.std(sample_data)}")
print(f"Standardized Mean (should be ~0): {np.mean(standardized_data)}")
print(f"Standardized Std Dev (should be ~1): {np.std(standardized_data)}")



fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

axes[0].hist(sample_data, bins=20, color='skyblue', edgecolor='black')
axes[0].axvline(np.mean(sample_data), color='red', linestyle='dashed', linewidth=2)
axes[0].set_title('Original Data Distribution (Mean shown in red)')
axes[0].set_xlabel('Raw Value')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.5)

# Standardized Data Histogram (Z-scores)
axes[1].hist(standardized_data, bins=20, color='lightgreen', edgecolor='black')
axes[1].axvline(np.mean(standardized_data), color='red', linestyle='dashed', linewidth=2)
axes[1].set_title('Standardized Data (Z-scores)')
axes[1].set_xlabel('Z-score (Standard Deviations)')
axes[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

After running the script:

The output confirms that after standardization, the mean of the dataset becomes very close to 0 and the standard deviation becomes very close to 1.

The plot on the left shows the original data centered around the raw mean (~70).

The plot on the right shows the exact same shape of the distribution, but the horizontal axis has been rescaled so that the center is at 0, and the units represent the number of standard deviations.
